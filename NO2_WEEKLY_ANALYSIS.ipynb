```python
from matplotlib.pyplot import axes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
air_quality=pd.read_csv("NO2.csv")
air_quality=air_quality.rename(columns={"utc":"datetime"})
air_quality
```

```python
print(list(air_quality.columns.values))
```

```python
print(air_quality["location"].unique())
```

```python
#type cast the date as datetime ,as before it was string
air_quality["datetime"]=pd.to_datetime(air_quality["datetime"])
```

```python
print("max is ")
print(air_quality["datetime"].max())
print("min")
print(air_quality["datetime"].min())
print("difference")
print(air_quality["datetime"].max()-air_quality["datetime"].min())
```

```python
#insert new column with name Month
air_quality["Month"]=air_quality["datetime"].dt.month
air_quality["week"]=air_quality["datetime"].dt.week
air_quality
```

```python
x=air_quality.groupby([air_quality["datetime"].dt.year,"location"])["value"].mean()
x
```

```python
#plot the data of year base 
fig,axs=plt.subplots(figsize=(30,14))
air_quality.groupby([air_quality["datetime"].dt.year,"location"])["value"].mean().plot(kind="bar",rot=0,ax=axs,stacked=True)
plt.xlabel("Year")
plt.ylabel("NO2")
```

```python
#plot the data of year base 
fig,axs=plt.subplots(figsize=(30,14))
air_quality.groupby([air_quality["datetime"].dt.month,"location"])["value"].mean().plot(kind="bar",rot=0,ax=axs,stacked=True)
plt.xlabel("month")
plt.ylabel("NO2")
```

```python
#plot the data of year base 
fig,axs=plt.subplots(figsize=(30,14))
air_quality.groupby([air_quality["datetime"].dt.week,"location"])["value"].mean().plot(kind="bar",rot=0,ax=axs,stacked=True)
plt.xlabel("week")
plt.ylabel("NO2")
```
