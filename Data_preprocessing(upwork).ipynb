```python

data_frame=pd.read_csv("output.csv",on_bad_lines='skip')
```

```python
import random
randon_values=['Yes','No','2','20',"ALL",'Happy',"TRUE","FLASE"]
randon_values_=['Yes','No','2','20',"ALL",'Happy',"TRUE","FLASE","1670","2001","Legend","maybe","0","0.5","Google","help me","Github",]
random_values_2=["None of above","all","both a and b","both c and b","both a and c"]
# from tables import Column
second_list=[]
third_list=[]
list_of=[]
num_to_select = 1                           # set the number to select here.
for i in range(0,data_frame.shape[0]):
  list_of_random_items = random.sample(randon_values, num_to_select)
  second_list_cloumns=random.sample(randon_values_, num_to_select)
  third=random.sample(random_values_2, num_to_select)
  third_list.append(third[0])
  second_list.append(second_list_cloumns[0])
  list_of.append(list_of_random_items[0])

third_final=[]
question=data_frame[["Question","Answer"]]

final_random=[]
second_list_final=[]
# question.insert(loc=2,column="option2",value=)
for i in range(0,data_frame.shape[0]):
  list_of_random_items = random.sample(list_of, num_to_select)
  second=random.sample(second_list, num_to_select)
  third=random.sample(third_list, num_to_select)
  third_final.append(third[0])
  second_list_final.append(second[0])
  final_random.append(list_of_random_items[0])
print(third_final)
print(final_random)
question.insert(loc=2,column="option3",value=final_random)
print(second_list_final)
question.insert(loc=2,column="option4",value=second_list_final)
question.insert(loc=2,column="option2",value=third_final)
```
